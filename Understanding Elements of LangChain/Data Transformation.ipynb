### ***Recursive Character Text Splitter***

```python
# Text Loader (Used for .txt files like "speech.txt")
from langchain_community.document_loaders import TextLoader

loader = TextLoader("speech.txt")
docs = loader.load()
print(docs)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Assuming 'docs' is a list of strings extracted from the WebBaseLoader
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)

# Split the content of each document
split_docs = text_splitter.split_documents(docs)

# 'split_docs' will now contain smaller chunks of text
for doc in split_docs:
    print(doc)
```

### ***Character Text Splitter***

```python
# Reading a PDF File
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Chapter.pdf")
docs = loader.load()
print(docs)
```

```python
from langchain.text_splitter import CharacterTextSplitter

# Assuming 'docs' is a list of strings extracted from the WebBaseLoader
text_splitter = CharacterTextSplitter(separator = "\n\n",
                                      chunk_size=500, 
                                      chunk_overlap=50)

# Split the content of each document
split_docs = text_splitter.split_documents(docs)

# 'split_docs' will now contain smaller chunks of text
for doc in split_docs:
    print(doc)
```
